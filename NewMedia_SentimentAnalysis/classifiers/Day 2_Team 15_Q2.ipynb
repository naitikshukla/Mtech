```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
newstopwords=stopwords.words("English") + ['the','is','it','may', 'was']
newstopwords.remove("not")
#WNlemma = nltk.WordNetLemmatizer()
tokenizer = RegexpTokenizer(r'[A-Za-z0-9_\']+')


def pre_process(text):
    tokens = nltk.word_tokenize(text)
    tokens = tokenizer.tokenize(text)
    tokens = [t.lower() for t in tokens]
#    tokens = [WNlemma.lemmatize(t) for t in tokens]
    tokens = [word for word in tokens if word not in newstopwords]
#    text_after_process = " ".join(tokens)
    return(tokens)

# Apply preprocessing to every document in the training set.
Review = testset['text']
testset['text'] = Review.apply(pre_process)
```

```python
test_pos = testset[(testset.Sentiment == 1)]
test_neg = testset[(testset.Sentiment == -1)]
```

```python
bigrams = test_pos['text'].apply(nltk.bigrams)
tot_bigrams = []
for i,value in bigrams.iteritems():
  tot_bigrams += list(value)
print(len(tot_bigrams))
FreqDist(tot_bigrams).most_common(50)
```

```python
trigrams = test_neg['text'].apply(nltk.trigrams)
tot_trigrams = []
for i,value in trigrams.iteritems():
  tot_trigrams += list(value)
print(len(tot_trigrams))
FreqDist(tot_trigrams).most_common(50)
```
